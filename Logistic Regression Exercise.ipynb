# Do the exercise in the Regression LE which is to discover if 4.8,2.5,5.3,2.4 is iris an virginica.
________

I have displayed all 4 attributes in the following regressions. The red line stands for an "iris virginica" and the green one for "non iris virginica":

>**sepal lenght 4.8   -> 100% Iris Virginica    
sepal width  2.5   -> ca. 70% Iris Virginica    
petal lenght 5.3   -> 100% Iris Virginica    
petal width  2.4   -> ca. 90% Iris Virginica**     

So the probability is very high, that the given data is an iris virginica!

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()

X = iris["data"][:, 3:]  # petal width
y = (iris["target"] == 2).astype(np.int)  # 1 if Iris-Virginica, else 0

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=42)
log_reg.fit(X, y)

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

plt.plot(X_new, y_proba[:, 1], "r-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "g--", linewidth=2, label="Not Iris-Virginica")
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
#It works!!! :)
plt.legend(loc="center left", fontsize=14)
```
